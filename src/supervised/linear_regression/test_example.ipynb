```python
from linear_regression import Model
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

# Dataset

```python
# Load dataset.
X, y = make_regression(random_state=312) 

# Use only one feature.
X = X[:, np.newaxis, 0]

# Split the data into train/test sets.
X_train = X[:-20] 
X_test = X[-20:] 

# Split the target into train/test sets.
y_train = y[:-20]
y_test = y[-20:]
```

# Model

```python
# Define model.
model1 = Model(normal_equation=True)
model2 = Model(epochs=10, lr=0.01)

# Fit dataset.
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)

# Make prediction.
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
```

# Display

```python
fig = plt.figure(figsize=(10,10))

fig.add_subplot(3, 1, 1)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred1, color='blue')

fig.add_subplot(3, 1, 2)
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred2, color='blue')

fig.add_subplot(3, 1, 3)
plt.plot(model2.training_errors)
plt.xlabel('Epochs')
plt.ylabel('Cost function')

plt.show()
```
